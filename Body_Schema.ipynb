# **Gain Modulation**

**What is Gain Modulation?**



The neurons with gain-modulatory effects are multisensory neurons, nonlinearly integrate two or more inputs together. One input is called driving input, and the other/s are called modulatory input which alters neurons sensitivity to the driving inputs.

**What is its function?**

These multisensory neurons basically integrates inputs together, and are found in different areas of the brain with different functionalites, namely coordinate transformation, variable-independant response, etc.


**Properties and Implementation**

First requirment for implementing gain modulation  properly is that the interaction mof driving and modulatory inputs have a multiplicative nature[1].

I implemented this multiplicative interaction in pyNN and Spinnaker5 using conductance based LIF model (IF_cond_exp ) in which incoming excitatory and inhibitory spikes change the conductance of the membrane potential , making it able to do the multiplicative intraction. Driving input are excitatory and modulating inputs were inhibitory.
